## CREAMOS LAS TABLAS E INSERTAMOS DATOS 

```python
import pandas as pd
import mysql.connector

# --- 1. DEFINICIÓN DE FUNCIONES ---

def crear_estructura_completa(config):
    print(">>> 1. Conectando para crear tablas...")
    try:
        conexion = mysql.connector.connect(**config)
        cursor = conexion.cursor()
        
        tablas = [
            "CREATE TABLE IF NOT EXISTS departments (department_id INT AUTO_INCREMENT PRIMARY KEY, department_name VARCHAR(100) UNIQUE NOT NULL);",
            "CREATE TABLE IF NOT EXISTS job_roles (job_role_id INT AUTO_INCREMENT PRIMARY KEY, role_name VARCHAR(100) UNIQUE NOT NULL);",
            "CREATE TABLE IF NOT EXISTS employees (employee_number INT PRIMARY KEY, age INT, gender VARCHAR(50), marital_status VARCHAR(50), education INT);",
            """CREATE TABLE IF NOT EXISTS employee_finance_details (
                finance_id INT AUTO_INCREMENT PRIMARY KEY, employee_number INT, monthly_income DECIMAL(10,2), 
                over_time TINYINT, department_id INT, job_role_id INT,
                FOREIGN KEY (employee_number) REFERENCES employees(employee_number) ON DELETE CASCADE,
                FOREIGN KEY (department_id) REFERENCES departments(department_id),
                FOREIGN KEY (job_role_id) REFERENCES job_roles(job_role_id));""",
            """CREATE TABLE IF NOT EXISTS employee_career_path (
                career_path_id INT AUTO_INCREMENT PRIMARY KEY, employee_number INT, years_at_company INT, 
                years_in_current_role INT, years_since_last_promotion INT, years_with_curr_manager INT, 
                total_working_years INT, FOREIGN KEY (employee_number) REFERENCES employees(employee_number) ON DELETE CASCADE);""",
            """CREATE TABLE IF NOT EXISTS employee_satisfaction (
                survey_id INT AUTO_INCREMENT PRIMARY KEY, employee_number INT, environment_satisfaction INT, 
                job_satisfaction INT, relationship_satisfaction INT, work_life_balance INT, 
                FOREIGN KEY (employee_number) REFERENCES employees(employee_number) ON DELETE CASCADE);"""
        ]
        
        for sql in tablas:
            cursor.execute(sql)
        
        conexion.commit()
        print(">>> PASO 1 COMPLETO: Estructura creada.")
        cursor.close()
        conexion.close()
    except Exception as e:
        print(f"!!! ERROR AL CREAR TABLAS: {e}")

def insertar_datos_totales(df, config):
    print(">>> 2. Iniciando inserción de datos...")
    df.columns = [col.lower().strip() for col in df.columns]
    
    try:
        conexion = mysql.connector.connect(**config)
        cursor = conexion.cursor()

        # Maestras e IDs
        depts = [(str(n),) for n in df['department'].unique() if pd.notna(n)]
        cursor.executemany("INSERT IGNORE INTO departments (department_name) VALUES (%s)", depts)
        
        roles = [(str(n),) for n in df['job_role'].unique() if pd.notna(n)]
        cursor.executemany("INSERT IGNORE INTO job_roles (role_name) VALUES (%s)", roles)
        conexion.commit()

        cursor.execute("SELECT department_id, department_name FROM departments")
        dict_depts = {name: id for id, name in cursor.fetchall()}
        cursor.execute("SELECT job_role_id, role_name FROM job_roles")
        dict_roles = {name: id for id, name in cursor.fetchall()}

        print(">>> Cargando filas (esto puede tardar un poco)...")
        for _, f in df.iterrows():
            emp_num = int(f['employee_number'])
            
            # 1. Employees
            cursor.execute("INSERT IGNORE INTO employees VALUES (%s, %s, %s, %s, %s)", 
                           (emp_num, int(f['age']), f['gender'], f['marital_status'], int(f['education'])))
            
            # 2. Finance
            cursor.execute("INSERT IGNORE INTO employee_finance_details (employee_number, monthly_income, over_time, department_id, job_role_id) VALUES (%s, %s, %s, %s, %s)",
                           (emp_num, float(f['monthly_income']), 1 if str(f['over_time']).lower() == 'yes' else 0, dict_depts.get(f['department']), dict_roles.get(f['job_role'])))

            # 3. Career Path
            cursor.execute("INSERT IGNORE INTO employee_career_path (employee_number, years_at_company, years_in_current_role, years_since_last_promotion, years_with_curr_manager, total_working_years) VALUES (%s, %s, %s, %s, %s, %s)",
                           (emp_num, int(f['years_at_company']), int(f['years_in_current_role']), int(f['years_since_last_promotion']), int(f['years_with_curr_manager']), int(f['total_working_years'])))

            # 4. Satisfaction
            cursor.execute("INSERT IGNORE INTO employee_satisfaction (employee_number, environment_satisfaction, job_satisfaction, relationship_satisfaction, work_life_balance) VALUES (%s, %s, %s, %s, %s)",
                           (emp_num, int(f['environment_satisfaction']), int(f['job_satisfaction']), int(f['relationship_satisfaction']), int(f['work_life_balance'])))

        conexion.commit()
        print(">>> PASO 2 COMPLETO: ¡Datos cargados con éxito!")
        cursor.close()
        conexion.close()
    except Exception as e:
        print(f"!!! ERROR AL INSERTAR: {e}")

# --- 2. CONFIGURACIÓN Y EJECUCIÓN ---

mi_config = {
    "host": "127.0.0.1",
    "user": "root",
    "password": "TU PASSWORD", # Reemplaza con tu pass
    "database": "corporation_abc"
}

# Cargar el CSV
try:
    df_insercion = pd.read_csv('empleados_abc_limpio.csv')
    print("CSV cargado correctamente.")
    
    # Llamar a las funciones
    crear_estructura_completa(mi_config)
    insertar_datos_totales(df_insercion, mi_config)
    
except FileNotFoundError:
    print("Error: No se encontró el archivo CSV. Asegúrate de que esté en la misma carpeta.")
except Exception as e:
    print(f"Error inesperado: {e}")
```
