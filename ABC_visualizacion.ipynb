# Visualización

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None) 
```

### El objetivo de este análisis es comprender los factores que influyen en la rotación de empleados dentro de la empresa.

## ¿Qué está pasando en la empresa?

### Mediante el analisis vamos a entender que esta fallando en la empresa y como podemos solucionarlo.

## % de empleados que se van

```python
df_visualizacion = pd.read_csv("empleados_limpio.csv")
```

```python
sns.countplot(x="attrition", data=df_visualizacion)
```

### Se ha analizado la variable **attrition** para comprender el nivel de rotación dentro de la empresa.

La mayoría de los empleados no abandonan la empresa (False)
Existe un grupo menor de empleados que sí abandonan (True)

### En porcentaje:

- 84% permanecen en la empresa
- 16% abandonan la empresa

Que quiere decir esto que aunque la mayoría de los empleados permanece en la empresa, existe un porcentaje significativo de rotación que debe ser analizado.

Vamos a ver que causas provocan este dato.



## ¿La gente se va por sueldo?

```python
sns.boxplot(x="attrition", y="monthly_income", data=df_visualizacion)
```

Los empleados que permanecen en la empresa (False) presentan una mediana salarial más alta (alrededor de 5.000).

Los empleados que abandonan la empresa (True) tienen una mediana salarial más baja (alrededor de 3.000).

Además, el grupo que permanece presenta:
Mayor dispersión salarial
Más valores altos (outliers)

Existe una relación clara entre salario y abandono:

Aunque existen algunos empleados con salarios altos que también abandonan, la tendencia general indica que los niveles salariales más bajos concentran mayor abandono.

## ¿cómo podriamos solucionarlo?

Revisar los rangos salariales más bajos

Analizar la competitividad salarial en el mercado

Implementar estrategias de retención para empleados con menor salario

## ¿se van los nuevos o los veteranos?

```python
sns.boxplot(x="attrition", y="years_at_company", data=df_visualizacion)
```

Se ha analizado la relación entre los años que un empleado lleva en la empresa y su abandono.

Los empleados que abandonan la empresa (True) presentan una menor antigüedad.

Mediana alrededor de 2–3 años

Los empleados que permanecen (False) tienen una mayor antigüedad:

Mediana alrededor de 5–6 años

Se observan algunos empleados con alta antigüedad que abandonan, aunque son casos menos frecuentes.

La rotación se concentra principalmente en los primeros años dentro de la empresa, lo que sugiere que el riesgo de abandono es mayor en fases iniciales.


## ¿cómo podriamos solucionarlo?

Mejorar procesos de onboarding

Implementar seguimiento en los primeros años

Desarrollar planes de crecimiento temprano


## ¿qué departamento tiene más fuga?

```python
sns.countplot(x="department", hue="attrition", data=df_visualizacion)
```

Research & Development presenta el mayor número de abandonos, aunque esto puede estar influido por su mayor volumen de empleados.

Por ello, es necesario analizar la proporción de abandono para determinar si realmente es el departamento con mayor riesgo.

```python
# proporción de abandono por departamento
df_visualizacion.groupby("department")["attrition"].mean()
```

Se ha analizado la proporción de empleados que abandonan la empresa en cada departamento.

- Sales → 20,27% 
- Human Resources → 19,05% 
- Research & Development → 14,05% 

El departamento de Sales presenta la mayor tasa de abandono, lo que indica un mayor riesgo de rotación en comparación con el resto de departamentos.

Aunque Research & Development presenta un mayor número absoluto de abandonos, su tasa relativa es menor, lo que sugiere una mayor estabilidad.

El principal foco de atención debería situarse en el departamento de Sales, donde el porcentaje de abandono es más elevado.

## ¿cómo podriamos solucionarlo?

Analizar condiciones laborales en Sales

Evaluar carga de trabajo y objetivos

Revisar políticas de incentivos

## Satisfacción con abandono

```python
sns.boxplot(x="attrition", y="job_satisfaction", data=df_visualizacion)
```

Los empleados que permanecen en la empresa (False) presentan niveles de satisfacción más altos:

Mediana alrededor de 3

Los empleados que abandonan la empresa (True) muestran niveles de satisfacción más bajos:

Mayor concentración en valores 1 y 2

Se observa una clara diferencia en la distribución entre ambos grupos.

Los empleados menos satisfechos tienen mayor probabilidad de abandonar la empresa.

## ¿cómo podriamos solucionarlo?

- Evaluar el clima laboral

- Implementar encuestas de satisfacción

- Desarrollar planes de mejora interna

- Fomentar reconocimiento y desarrollo profesional


Nos damos cuenta como el salario + satisfaccion en el trabajo son claves de abandono.

## Horas extras con abandono

```python
sns.countplot(x="over_time", hue="attrition", data=df_visualizacion)
```

- Los empleados que no realizan horas extra (False) presentan un menor número de abandonos.

-  empleados que sí realizan horas extra (True) muestran un número significativamente mayor de abandonos en proporción.

-  empleados que realizan horas extra tienen mayor probabilidad de abandonar la empresa.

## ¿cómo podriamos solucionarlo?

Revisar la carga de trabajo en equipos con horas extra

Implementar medidas de conciliación laboral

Controlar el exceso de horas extra

Evaluar impacto en el bienestar del empleado

### Hasta ahora podemos ver causas de abandono claras:

Salario bajo → abandono

Baja satisfacción → abandono

Horas extra → abandono

## Formación con abandono

```python
sns.boxplot(x="attrition", y="training_times_last_year", data=df_visualizacion)
```

Se ha analizado la relación entre el número de formaciones realizadas en el último año y el abandono de empleados.

Tanto los empleados que permanecen (False) como los que abandonan (True) presentan distribuciones muy similares.

La mediana en ambos casos se sitúa aproximadamente en 2–3 formaciones al año.

El número de formaciones realizadas no es un factor determinante en el abandono de empleados en este caso.
