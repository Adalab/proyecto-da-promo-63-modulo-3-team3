¿Por qué me salen los nulos 0 si habia 44?

Cambiamos el tipo de dato a boleano incorrectamente anteriormente en la parte del código, y por eso desaparecieron los yes/no:

#CONVIERTO LOS VALORES DE ESTAS DOS COLUMNAS A BOOLEANOS
columnas_bool = ["over_time", "attrition"]
for col in columnas_bool:
    df_empleados[col] = df_empleados[col].astype("bool")


y también la posibilidad de ver los 44 nulos reales.

En over_time teniamos valores de texto, algo como:

"yes", "no" y nulos.

pasa esto:
cualquier texto no vacío → True y los NaN en este contexto también te terminan quedando como True.
O sea, la columna perdió su contenido original y quedó convertida en booleanos.

Vamos a recuperar esa columna, con el archivo original.

El código que teniamos que utilizar para pasar over_time a boleano seria:

df_empleados["over_time"] = df_empleados["over_time"].map({"yes": True, "no": False})


```python
# Reconstruimos la columna 
df_original = pd.read_csv("hr.csv")
df_empleados["over_time"] = df_original["OverTime"]
df_empleados["attrition"] = df_original["Attrition"]
```

```python
# Homogenizamos columnas
df_empleados["over_time"] = (
    df_empleados["over_time"]
    .str.replace(r'([a-z])([A-Z])', r'\1_\2', regex=True)
    .str.lower()
    .str.strip()
)

df_empleados["attrition"] = (
    df_empleados["attrition"]
    .str.replace(r'([a-z])([A-Z])', r'\1_\2', regex=True)
    .str.lower()
    .str.strip()
)
```

```python
# Cambiamos el tipo de dato de la columna 'over_time' de forma correcta.
df_empleados["over_time"] = df_empleados["over_time"].map({"yes": True, "no": False})
df_empleados["attrition"] = df_empleados["attrition"].map({"yes": True, "no": False})
```

```python
# Comprobamos los nulos
df_empleados["over_time"].isnull().sum()
```
