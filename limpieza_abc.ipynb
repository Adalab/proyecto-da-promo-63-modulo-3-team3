```python
# ── Mapa de calor con máscara (triángulo superior derecho oculto) ──────────
def plot_heatmap_correlacion(df_numericas1, figsize=(30, 15), cmap="mako"):
    """
    Muestra la matriz de correlación como mapa de calor.
    Oculta el triángulo superior derecho (duplicidad) con una máscara.

    Parámetros
    ----------
    df_numericas : DataFrame con columnas numéricas
    figsize : tamaño del lienzo (ancho, alto)
    cmap : paleta de colores de seaborn/matplotlib
    """
    # 1. Matriz de correlación
    matrix = df_numericas1.corr()

    # 2. Máscara: True en el triángulo superior (se ocultará)
    mascara = np.triu(np.ones_like(matrix, dtype=bool))

    # 3. Lienzo
    plt.figure(figsize=figsize)

    # 4. Heatmap
    sns.heatmap(
        matrix,
        annot=True,          # muestra los valores numéricos
        fmt=".2f",          # dos decimales
        cmap=cmap,          # paleta de color
        vmin=-1, vmax=1,    # escala completa del coeficiente
        mask=mascara,        # oculta triángulo superior derecho
        linewidths=0.5,     # separa visualmente las celdas
        square=True         # celdas cuadradas, más limpio
    )

    plt.title("Matriz de correlación — variables numéricas", pad=14)
    plt.tight_layout()
    plt.show()


# ── Llamada ───────────────────────────────────────────────────────────────
plot_heatmap_correlacion(df_numericas1)
```

📊 Resumen de la matriz de correlación
La mayoría de variables tienen poca relación entre sí (correlaciones cercanas a 0).
🔑 Relaciones clave
Salario y nivel del puesto → relación muy alta
A mayor nivel, mayor salario.
Edad y experiencia → relación moderada
Más edad implica más años trabajados.
Tiempo en la empresa → variables relacionadas entre sí
Los años en la empresa, en el puesto y con el jefe están conectados.
Desempeño y aumento salarial → relación alta
Mejor rendimiento → mayor aumento de sueldo.
🧠 Conclusión

```python
#vemos los nulos por columna
print(df_empleados.isnull().sum())

print("-" * 100)

print(df_empleados.select_dtypes(include="number").isnull().sum())

print("-" * 100)

print(df_empleados.select_dtypes(include="str").isnull().sum())
```

```python
#calculo el portentaje de nulos 
nulos_cate_empleados = (df_empleados.select_dtypes(include="str").isnull().sum() / df_empleados.shape[0]) * 100

nulos_cate_empleados = nulos_cate_empleados.reset_index()

nulos_cate_empleados.rename(columns={"index":"nombre_columnas", 0: "%_nulos"},inplace=True)

nulos_cate_empleados
```

```python
mascara_nulos = nulos_cate_empleados["%_nulos"] > 0

nulos_cate_empleados = nulos_cate_empleados[mascara_nulos]

nulos_cate_empleados
```

```python
lista_categorias_nulos = nulos_cate_empleados["nombre_columnas"].values

for col in lista_categorias_nulos:
    print(f"ESTAMOS ANALIZANDO LA CATEGORIA {col}")
    print(f"Los recuentos para cada categoria son {df_empleados[col].value_counts(normalize=True) * 100}")
    print("=" * 100)
```

```python
for col in df_empleados.select_dtypes(include='str').columns:
    if not df_empleados[col].mode().empty:
        valor_frecuente = df_empleados[col].mode()[0]
        df_empleados[col] = df_empleados[col].fillna(valor_frecuente)
```
